```python
import pygly2
from pygly2.plot import plot
import matplotlib
%matplotlib inline

matplotlib.rcParams['figure.figsize'] = 10, 8


broad_n_glycan = pygly2.io.glycoct.loads('''
RES
1b:b-dglc-HEX-1:5
2s:n-acetyl
3b:b-dglc-HEX-1:5
4s:n-acetyl
5b:b-dman-HEX-1:5
6b:a-dman-HEX-1:5
7b:b-dglc-HEX-1:5
8s:n-acetyl
9b:b-dgal-HEX-1:5
10b:a-lgal-HEX-1:5|6:d
11b:b-dglc-HEX-1:5
12s:n-acetyl
13b:b-dgal-HEX-1:5
14b:a-dman-HEX-1:5
15b:b-dglc-HEX-1:5
16s:n-acetyl
17b:b-dgal-HEX-1:5
18b:b-dglc-HEX-1:5
19s:n-acetyl
20b:b-dgal-HEX-1:5
LIN
1:1d(2+1)2n
2:1o(4+1)3d
3:3d(2+1)4n
4:3o(4+1)5d
5:5o(3+1)14d
6:5o(6+1)6d
7:6o(2+1)11d
8:6o(6+1)7d
9:7d(2+1)8n
10:7o(3+1)10d
11:7o(4+1)9d
12:11d(2+1)12n
13:11o(4+1)13d
14:14o(2+1)18d
15:14o(4+1)15d
16:15d(2+1)16n
17:15o(4+1)17d
18:18d(2+1)19n
19:18o(4+1)20d''').next()
```

Draw the tree and label each node

```python
draw_tree, ax = plot(broad_n_glycan)

def unspool(draw_tree):
    yield draw_tree
    for child in draw_tree:
        for descend in unspool(child):
            yield descend

nodes = list(unspool(draw_tree))
for node in nodes:
    x, y = node.coords('h')
    ax.text(x, y + 0.2, node.tree.id, color='red')
```

```python
def pooler(gen):
    while True:
        yield gen.next(), gen.next()
matplotlib.rcParams["figure.figsize"] = 16, 12
for substructure, fragment in zip(broad_n_glycan.clone().substructures(max_cleavages=1), pooler(broad_n_glycan.clone().fragments('BY'))):
    y_frag, b_frag = fragment
    fig, ax = matplotlib.pyplot.subplots(1, 2)
    dt, sub_ax = plot(substructure.parent_tree, ax = ax[0], center=True)
    sub_ax.set_title(str(y_frag) + r" $\to$ " + broad_n_glycan.name_fragment(y_frag))
    dt, sub_ax = plot(substructure.child_tree, ax=ax[1], center=True)
    sub_ax.set_title(str(b_frag) + r" $\to$ " + broad_n_glycan.name_fragment(b_frag) + '\n')    
```
